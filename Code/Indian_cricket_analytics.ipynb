```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from moviepy.editor import VideoFileClip, AudioFileClip
from matplotlib.offsetbox import OffsetImage, AnnotationBbox


logo_path = '/content/Designer.png'
logo = plt.imread(logo_path)

# Load the CSV file
file_path = '/content/runs.csv'
data = pd.read_csv(file_path)

# Extract AGE column as x-axis
x = data['AGE']

# Extract player columns
player_columns = data.columns[1:]

# Set a style for the plot
plt.style.use("seaborn-v0_8-darkgrid")

# Adjust the figure size to 16:9 aspect ratio
fig, ax = plt.subplots(figsize=(16, 9))

# Initialize empty lines for each player with different styles
line_styles = ['-', '--', '-.', ':', '-', '--', '-.', ':']
lines = {player: ax.plot([], [], label=player, linestyle=line_styles[i % len(line_styles)], linewidth=2)[0] for i, player in enumerate(player_columns)}

# Highlight the current age
age_text = ax.text(0.1, 0.6, '', transform=ax.transAxes, fontsize=12, bbox=dict(facecolor='grey', alpha=0.5))

# Set up plot limits and labels
ax.set_xlim(x.min(), x.max())
ax.set_ylim(data[player_columns].min().min(), data[player_columns].max().max())
ax.set_xlabel('Age', fontsize=14)
ax.set_ylabel(' Runs', fontsize=14)
ax.set_title('Indian Cricket Legends\' Age vs Runs in ODIs', fontsize=11)
imagebox = OffsetImage(logo, zoom=0.05)  # Adjust zoom factor as needed
ab = AnnotationBbox(imagebox, (1, 1), xycoords='axes fraction', frameon=False)
ax.add_artist(ab)

# Ensure all x-axis ticks are shown
ax.set_xticks(x)

# Customize background color
fig.patch.set_facecolor('#ffffff')  # Neon green example
ax.set_facecolor('#ebecf0')

def update(frame):
    for player in player_columns:
        lines[player].set_data(x[:frame+1], data[player][:frame+1])
    age_text.set_text(f'Age: {x.iloc[frame]}')
    return list(lines.values()) + [age_text]

# Create animation with 2-second interval per frame
ani = animation.FuncAnimation(fig, update, frames=len(x), interval=2000, blit=True)

# Allow you to change the labels easily
player_labels = {player: player.replace('_', ' ') for player in player_columns}

# Update the legend labels
for player in player_columns:
    lines[player].set_label(player_labels[player])

# Adjust layout to ensure legend is not cropped
plt.subplots_adjust(left=0.05, right=0.95, top=0.90, bottom=0.25)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=5, fontsize=10)

# Save the animation with updated labels
output_labeled_video_path = '/content/line_chart_animation_with_labels.mp4'
ani.save(output_labeled_video_path, writer='ffmpeg', fps=0.5)

# Add music to the animation
audio_file_path = '/content/music.mp3'  # Replace with the path to your music file

# Load video and audio clips
video_clip = VideoFileClip(output_labeled_video_path)
audio_clip = AudioFileClip(audio_file_path).subclip(0, video_clip.duration)

# Ensure the final clip dimensions are the same as the original
final_clip = video_clip.set_audio(audio_clip)

# Save the final video with audio
final_output_path = '/content/line_chart_animation_with_music.mp4'
final_clip.write_videofile(final_output_path, codec='libx264', audio_codec='aac')

print("Final video saved at:", final_output_path)
```
